# Phân tích chung

```python
# Thương hiệu bán chạy nhất
top_brands = df['Trademarks'].value_counts().head(10)
print("Top 10 thương hiệu có sản phẩm được bày bán:")
print(top_brands,'\n')

# Giá trung bình của mỗi thương hiệu
average_price_by_brand = df.groupby('Trademarks')['Product Price'].mean().sort_values(ascending=False).head(10)
print("Giá cả trung bình theo thương hiệu (VND):")
print(average_price_by_brand,'\n')

# Sự khác biệt về giá giữa các nhóm tuổi
average_price_by_age = df.groupby('Age Usage')['Product Price'].mean()
print("Giá cả trung bình theo độ tuổi sử dụng (VND):")
print(average_price_by_age,'\n')
# Sự khác biệt về giá giữa các nhóm tuổi
average_price_by_cate = df.groupby('Catergory')['Product Price'].mean()
print("Giá cả trung bình theo nhóm mặt hàng (VND):")
print(average_price_by_cate)
```

```python
average_price_by_brand = average_price_by_brand.sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=top_brands.index, y=top_brands.values)
plt.title('Top 10 thương hiệu có sản phẩm được bày bán')
plt.xlabel('Thương hiệu')
plt.ylabel('Số lượng')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=average_price_by_brand.index, y=average_price_by_brand.values)
plt.title('Giá cả trung bình theo thương hiệu')
plt.xlabel('Thương hiệu')
plt.ylabel('Giá trung bình')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=average_price_by_age.index, y=average_price_by_age.values)
plt.title('Giá cả trung bình theo tuổi')
plt.xlabel('Độ tuổi')
plt.ylabel('Giá trung bình')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
sns.barplot(x=average_price_by_cate.index, y=average_price_by_cate.values)
plt.title('Giá cả trung bình theo nhóm mặt hàng')
plt.xlabel('Mặt hàng')
plt.ylabel('Giá trung bình')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

## Phân tích giá cả 

```python
print('Giá trị trung bình: ',df['Product Price'].mean())
print('Độ lệch chuẩn:', df['Product Price'].std())
```

```python
from scipy import stats
mean_ci = stats.norm.interval(0.95, loc=df['Product Price'].mean(), scale=df['Product Price'].std() / np.sqrt(len(df)))

# Tính khoảng tin cậy cho độ lệch chuẩn (ví dụ: 95% khoảng tin cậy)
std_ci = stats.norm.interval(0.95, loc=df['Product Price'].mean(), scale=df['Product Price'].std())

# In ra khoảng tin cậy cho giá trung bình và độ lệch chuẩn
print("95% Khoảng tin cậy cho giá trung bình:", mean_ci)
print("95% Khoảng tin cậy cho độ lệch chuẩn:", std_ci)
```

```python
mean_price = df['Product Price'].mean()
std_dev_price = df['Product Price'].std()
```

```python
plt.figure(figsize=(10, 5))
plt.scatter(range(len(df)),df['Product Price'], label='Prices', color='blue')
plt.axhline(mean_price, color='red', linestyle='--', label=f'Mean: {mean_price:.2f}')
plt.fill_between(range(len(df)), mean_price - std_dev_price, mean_price + std_dev_price, 
                 color='red', alpha=0.2, label=f'Std Dev: ±{std_dev_price:.2f}')

plt.title('Giá với trung bình và độ lệch chuẩn')
plt.xlabel('Trị số')
plt.ylabel('Giá')
plt.legend()
plt.grid(True)
plt.show()
```

```python
cheap = df[df['Product Price'] < 150000.0].shape[0]
medium = df[(df['Product Price'] >= 150000.0) & (df['Product Price'] <= 500000.0)].shape[0]
expensive = df[df['Product Price'] > 500000.0].shape[0]

# Tính phần trăm
sizes = [cheap / len(df) * 100, medium / len(df) * 100, expensive /len(df)* 100]
labels = ['Thấp', 'Trung Bình và Cận Cao Cấp', 'Cao Cấp']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Màu sắc tương ứng
explode = (0.1, 0.1, 0)  # Đẩy miếng "Thấp" ra ngoài

# Vẽ biểu đồ tròn
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)

plt.axis('equal')  # Đảm bảo biểu đồ tròn không bị biến dạng
plt.title('Phân phối sản phẩm theo mức giá')
plt.show()
```

```python
cheap = df[df['Age Usage'] == 'Thời kỳ răng sữa'].shape[0]
medium = df[df['Age Usage'] == 'Giai đoạn thiếu niên'].shape[0]
expensive = df[df['Age Usage'] == 'Người lớn'].shape[0]

sizes = [cheap / len(df) * 100, medium / len(df) * 100, expensive /len(df)* 100]
labels = ['Trẻ sơ sinh', 'Trẻ vị thành niên', 'Đồ cho mẹ']
colors = ['#1f77b4', '#ff7f0e', '#2ca02c']  # Màu sắc tương ứng
explode = (0.1, 0.1, 0.3)  # Đẩy miếng "Thấp" ra ngoài

# Vẽ biểu đồ tròn
plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%',
        shadow=True, startangle=140)

plt.axis('equal')  # Đảm bảo biểu đồ tròn không bị biến dạng
plt.title('Phân phối sản phẩm theo nhóm tuổi sử dụng')
plt.show()
```

```python
brand_sales = df['Trademarks'].value_counts(normalize=True) *100

# Sắp xếp các thương hiệu theo thứ tự giảm dần của số lượng bán
top_brands = brand_sales.head(10).index.tolist()

# Lấy thông tin về giá và tuổi sử dụng của các thương hiệu được chọn
top_brands_info = df[df['Trademarks'].isin(top_brands)]
```

```python
top_brands_info
```

```python
plt.figure(figsize=(10, 6))
ax = sns.countplot(data=top_brands_info, x='Trademarks', order=top_brands)

# Tính toán tỷ lệ phần trăm và hiển thị trên biểu đồ
total = len(top_brands_info)
for p in ax.patches:
    percentage = '{:.1f}%'.format(100 * p.get_height() / total)
    x = p.get_x() + p.get_width() / 2 - 0.05
    y = p.get_height()
    ax.annotate(percentage, (x, y), ha='center')

# Tùy chỉnh biểu đồ
plt.title('Tỷ lệ của 10 thương hiệu hàng đầu')
plt.xlabel('Thương hiệu')
plt.ylabel('Số lượng')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```

```python
plt.figure(figsize=(10, 6))
box_plot = sns.barplot(x='Trademarks', y='Product Price', data=top_brands_info, hue='Age Usage', palette='Set2')

# Tùy chỉnh biểu đồ
plt.title('Biểu đồ phân phối giá cả theo thương hiệu và độ tuổi sử dụng')
plt.xlabel('Thương hiệu')
plt.ylabel('Giá')
plt.xticks(rotation=45)
plt.grid(True)
plt.legend(title='Độ tuổi sử dụng')
plt.tight_layout()

# Hiển thị biểu đồ
plt.show()
```
